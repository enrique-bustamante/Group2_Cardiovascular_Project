```python
# Add the Pandas dependency.
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
print('x' in np.arange(5))   #returns False, without Warning
```

```python
#Load the CSV
cardio_df = pd.read_csv("cardio_train.csv", index_col= "id")
cardio_df
```

```python
#Determine if there are any missing values
cardio_df.count()
```

```python
#Determine the data types for the cardio DF
cardio_df.dtypes
```

```python
#Change age column from days to years
cardio_df['age'] = cardio_df['age'] / 365
cardio_df.head()
```

```python
#Format age column
cardio_df['age'] = cardio_df['age'].map("{:,.0f}".format)
cardio_df
```

```python
#Create DF for cardiovascular patients
cardio_df_yes = cardio_df.loc[cardio_df["cardio"] == "1"]
cardio_df_yes
```

```python
#Create DF for non heart disease patients
cardio_df_no = cardio_df.loc[cardio_df["cardio"] == "0"]
cardio_df_no
```
